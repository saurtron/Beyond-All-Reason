```python
# Automatically generates sounds.lua file which maps filenames to lengths in seconds
# Author Beherith mysterme@gmail.com


import struct
import os

def get_wav_length(fname):
    with open(fname,'rb') as f:
        data = f.read(40)
        length = struct.unpack('<I', data[4:8])[0] - 36
        channels = struct.unpack('<H', data[22:24])[0]
        samplerate = struct.unpack('<I', data[24:28])[0]
        bitspersample = struct.unpack('<H', data[34:36])[0]
        return length / (channels*samplerate * (bitspersample/8))
    
luaf = open('sound_file_lengths.lua','w')
luaf.write("--Automatically generated by sound_file_lengths.ipynb, do not edit!\n\n")
luaf.write('local wavLengths = {\n')
for root, dir, files in os.walk(os.getcwd()):
    for file in files:
        print(root,dir,file)
        abspath = os.path.join(root[len(os.getcwd()) + 1:], file)
        if abspath.lower().endswith('.wav'):
            wavlen = get_wav_length(abspath)
            print (abspath, wavlen)
            luaf.write('\t["%s"] = %f,\n'%(abspath.replace('\\', '/'), wavlen))
luaf.write('}\nreturn wavLengths\n')
luaf.close() 
        
```
